```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv('NIGERIA_DEMOGRAPHICS_DATA.csv')
data.head()
```

```python
#To call all columns since it wasnt all that appeared in the data set
pd.set_option('display.max_columns', None)
data.head()
```

```python
#Renaming columns because they look too lengthy in words
data.columns = ['date', 'Nigerian citzen', 'gender', 'age group', 'married', 'parenting', 'location', 'region','occupation', 'job_satiated', 'primary income source?','monthly salary range', 'is salary family sufficient?', 'other income source','tax compliant?', 'largest expense', 'dream property', 'other income earnings', 'monthly expenses', 'can Nigeria become better', 'describe government', 'email']
data.head()
```

```python
#To rearrange columns
cols = list(data.columns.values)
data = data[cols[0:11] + cols[12:17] + [cols[11]] + cols[17:22]]
data.head()
```
